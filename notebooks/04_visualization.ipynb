```python
import sys
import os
import yaml

# =========================
# Move to project root
# =========================

os.chdir("..")

# Add project path
sys.path.append(os.getcwd())

print("Current Directory:", os.getcwd())

# =========================
# Import Modules
# =========================

from src.data_loader import get_latest_file, load_data
from src.data_cleaning import clean_data
from src.outlier_detection import detect_outliers_iqr
from src.visualization import generate_charts

# =========================
# Load Config
# =========================

with open("config/config.yaml", "r") as file:
    config = yaml.safe_load(file)

# =========================
# Get Latest File
# (Use archive folder because pipeline moved raw file)
# =========================

raw_path = "data/archive"

file = get_latest_file(raw_path)

print("Using file:", file)

# =========================
# Load Data
# =========================

df = load_data(file)

# =========================
# Clean Data
# =========================

cleaned_df = clean_data(df, config)

# =========================
# Remove Outliers
# =========================

final_df = detect_outliers_iqr(
    cleaned_df,
    config
)

print("Final Data Shape:", final_df.shape)

# =========================
# Generate Charts
# =========================

generate_charts(
    final_df,
    "outputs/charts"
)

print("🎯 Visualization Completed Successfully")

```
